# Selecting the relevant colums

```python
df=df[['eventid', 'iyear', 'imonth', 'country', 'region','provstate','city','crit1', 'crit2', 'crit3','success', 'suicide', 'attacktype1','targtype1','natlty1','gname','guncertain1','claimed','weaptype1','nkill','nwound']]
```

```python
df.head()
```

```python
df.isnull().any()
```

# Removing Null Values

```python
df = df.dropna()
```

```python
df.isnull()
```

# As the null values have been eliminated, we can start the exploratory data analysis

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.countplot(x='success',data=df)
plt.xlabel('no. of successful attacks')
plt.ylabel('No.of attacks')
plt.title('Successful Terrorist attacks over the years')
```

```python
plt.figure(figsize=(31,20),dpi=400)
sns.barplot(df['city'].value_counts()[:15].index,df['city'].value_counts()[:15].values)
plt.title("Top 10 areas that are attacked")
plt.xlabel("Area")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(31,20),dpi=500)
sns.countplot(x='iyear',data=df,color='red')
plt.xlabel('year')
plt.ylabel('number of attacks')
plt.title('terrorist attacks over the years')
```

```python
df['city'].value_counts().nsmallest(10).plot(kind='bar',figsize=(10,5), color='green')
```
