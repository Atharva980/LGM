```python
import matplotlib.pyplot as plt
image = cv2.imread('tom.jpg')
```

```python

image1 = plt.imshow(image[...,::-1])
image1
```

```python
image_to_grayscale = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
```

```python

plt.imshow(image_to_grayscale,  cmap='gray' )
```

```python

image2 = 255-image_to_grayscale
```

```python
plt.imshow(image2, cmap='gray')
```

```python

image3 = cv2.blur(image2,(100,100),cv2.BORDER_DEFAULT)
```

```python
plt.imshow(image3, cmap='gray')
```

```python

image3 = cv2.GaussianBlur(image2,(5,5),0)
```

```python

plt.imshow(image2,cmap='gray')
```

```python
image4 = 255-image3
```

```python
imag = cv2.divide(image_to_grayscale, image4)
```

```python
plt.imshow(imag, cmap='gray')
```
